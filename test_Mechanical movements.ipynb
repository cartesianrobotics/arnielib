# Repeatedly moving up and down to see if step would miss

```python
ar.move(x=400, y=200)
```

```python
speed = 2000
for i in range(100):
    ar.move(z=100, speed_z=speed)
    time.sleep(0.1)
    ar.move(z=0, speed_z=speed)
    print ("step %s", i)
```

1. skipped steps once in 70 repeats
2. Switched to 1/2 microstepping. No skipped steps after 100 repeats
3. Attached p1000. No skipped steps
4. Moved Z to 400, moved between 400 and 500 for 100 times. No skipped steps
5. Moved 0-10 200 times. No skipped steps
6. Increased speed to 4000. Steps skipped on the way up every time.
7. Speed 3000. No skipped steps on every test.
