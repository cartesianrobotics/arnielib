## pipettor basic operations

```python
a.set_plunger_level(pipettor=ar.current_tool_device, level=20)
```

```python
# Speed from 0 to 1000 for 1000 and 200 pipettors; from 0 to 400 for 20 uL pipettor
# Default is the last speed set for the device

a.set_pipettor_speed(ar.current_tool_device, speed=700)
```

```python
a.set_plunger_level(pipettor=ar.current_tool_device, level=2)
```

```python
ar.current_tool_device.write("$$")
```

```python
ar.current_tool_device.write_wait("$$")
```

```python
# Moving servo
ar.current_tool_device.write_wait("M5")
```

```python
# Servo disengaged (will not drop tip)
ar.current_tool_device.write_wait("M3 S90")
```

```python
# Engage servo (moving will drop tip)
ar.current_tool_device.write_wait("M5")
```

```python
# drop tip
ar.current_tool_device.drop_tip()
```

## Calibrate pipettor

```python
# x_n, y_n - stalagmite coordinates
# initial_point - height from which to start calibration, to speed up
a.calibrate_pipettor_tip(robot=ar, x_n=0, y_n=0)
```

## Taking a tip

```python
ar.move(z=3000)
```

```python
a.pickup_tip(robot=ar, x_n=0, y_n=2, well_x_n=0, well_y_n=1)
```

# Pipetting from slot 0,1 to 0,0

```python
a.approach_well(robot=ar, x_n=1, y_n=3, well_x_n=1, well_y_n=0)
```

```python
# After stalaktite re-calibration, we need also to re-calibrate everything on the floor
# Will be fixed
```

```python
# expected_liquid_level - mm from tube bottom to liquid level expected after finishing pipetting
# plunger_level - level at which plunger will go down before immersing into liquid
# After going into the liquid, plunger moves to 0
# speed - speed of plunger movement. Default 700
# delay - when plunger finishes moving up, wait for this seconds
a.uptake_liquid(robot=ar, x_n=1, y_n=3, expected_liquid_level=10, plunger_level=30, speed=700, delay=7)
```

```python
# Now pipetting into the other well
a.approach_well(robot=ar, x_n=1, y_n=3, well_x_n=0, well_y_n=0)
```

```python
# Pipette down
a.release_liquid(robot=ar, x_n=1, y_n=3, plunger_level=30, speed=700, delay=5)
```
